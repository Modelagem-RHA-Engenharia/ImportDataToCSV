```
from google.colab import files
import io

uploaded  = files.upload()
file_name = next(iter(uploaded))
file_name
```

```
import pandas as pd
from datetime import datetime as dt

gfile_name = io.StringIO(uploaded[file_name].decode('UTF-8').strip())

# Nome das colunas
_columms_names = [
                  'ID',
                  'latitude',
                  'longitude',
                  'id_subbacia',
                  'AreaDREN',
                  'AreaCEL',
                  'Hmx',
                  'Hmn',
                  'length',
                  'slope',
                  'ID_CelJUS',
                  '_aux'
                 ] 
# Tipo dos dados da coluna
_dtype = {
          'ID': int,
          'latitude': float,
          'longitude': float,
          'id_subbacia': int,
          'AreaDREN': float,
          'AreaCEL': float,
          'Hmx': int,
          'Hmn': int,
          'length': float,
          'slope': float,
          'ID_CelJUS': int,
          '_aux': int
         }

df = pd.read_csv(gfile_name, 
                 # Se possui cabeçalho
                 header=None, 
                 # Tipo de delimitador
                 delimiter=r"\s+",
                 # Nome das colunas
                 names= _columms_names,
                 # Tipo dos dados da coluna
                 dtype=_dtype
                 )
df
```

```
df.info()
```

```
df.to_csv('file.csv', sep=';', encoding='utf-8')
files.download('file.csv')
```
